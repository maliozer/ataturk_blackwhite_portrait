```
originalImage = cv2.imread('./ata_colored.jpg')
```

####conver the image to gray scale in order to apply threshold operation

```
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
```

####applying binary threshold to grayscaled image

```
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255, cv2.THRESH_BINARY)
```

####comparison between original and output images

```
#importing matplotlib
from matplotlib import pyplot as plt

#comparison between original and output images
plt.imshow(cv2.cvtColor(originalImage, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(grayImage, cmap='gray')
plt.show()

plt.imshow(blackAndWhiteImage, cmap='gray')
plt.show()
```
